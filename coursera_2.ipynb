# Распределение Фишера

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import scipy.stats as sts
d1=100 #Два параметра для распределения фишера
d2=100
f_rv = sts.f(d1,d2)
sample = f_rv.rvs(1000)

```

 # Строим гистограмму по сгенерированной выборке объемом 1000 и ее теоретическую плотность

```python
plt.hist(sample,density=True)
x = np.linspace(0,4,1000)
pdf_f = f_rv.pdf(x)
plt.plot(x,pdf_f, label='theoretical pdf', alpha=0.5)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

# гистограммы распределений выборочных средних при н равно 5

```python
sample_averages=[]
n=5
for i in range(1,1001):
    sample=f_rv.rvs(n)
    temp_average=sum(sample)/n
    sample_averages.append(temp_average)

plt.hist(sample_averages,density=True)
x = np.linspace(0,2,100)
math_exp=d2/(d2-2)
disp=(2*(d2**2)*(d1+d2-2))/(d1*(d2-4)*((d2-2)**2))
print(math_exp,disp)
plt.plot(x,sts.norm(math_exp,disp).pdf(x) , lw=1,
             label='$\mathcal{N}' + '({}, {})$'.format(0, 1))
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
